```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread(r'C:\Users\julia\Documents\masters\deep_learning\text_recognition\data\train\train\complex\rus_00190_C226OC35.png')
img = img[:, :, 0]
ret, th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)

ada_mean_thr = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
ada_gaus_thr = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 127)', 'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, ada_mean_thr, ada_gaus_thr]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
